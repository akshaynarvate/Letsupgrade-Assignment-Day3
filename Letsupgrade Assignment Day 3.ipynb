```python
x1 = 3 * np.random.rand(10, 1)
y1 = 4 + 3 *x1* np.random.randn(10, 1)

x2 = 2 * np.random.rand(10, 1)
y2 = 2 + 4 *x1* np.random.randn(10, 1)

dataset = [(x1, y1),
           (x2, y2)]

plt.scatter(dataset[0][1], dataset[0][1], c="red", s=20,)
plt.scatter(dataset[1][0], dataset[1][1], c="green", s=100, label="From X2, Y2")
plt.title("Labelled data")
plt.xlabel("$X axis$")
plt.ylabel("$Y axis$")
# For enabling the labelling
plt.legend()
plt.show()
```

```python
# 4) Apply simple linear regression and submit it as an assignment.

import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
x = np.array([5, 15, 25, 35, 45, 55]).reshape((-1, 1))
y = np.array([5, 20, 14, 32, 22, 38])
```

```python
print(x)
```

```python
print(y)
```

```python
model = LinearRegression()
```

```python
model.fit(x, y)
```

```python
r_sq = model.score(x, y)
print('coefficient of determination:', r_sq)
```

```python
print('intercept:', model.intercept_)

print('slope:', model.coef_)
```

```python
y_pred = model.predict(x)
print('predicted response:', y_pred, sep='\n')
```

```python
y_pred = model.intercept_ + model.coef_ * x
print('predicted response:', y_pred, sep='\n')
```
